```python
import pandas as pd
import numpy as np
sleep_eff=pd.read_csv("Sleep_Efficiency.csv")
sleep_eff.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cols=['ID','Age'	,'Gender'	,'Bedtime'	,'Wakeup time',	'Sleep duration',	'Sleep efficiency'	,'REM sleep percentage'	,'Deep sleep percentage'	,'Light sleep percentage',	'Awakenings','Caffeine consumption'	,'Alcohol consumption'	,'Smoking status','Exercise frequency']
for f in cols:
    sleep_eff[f]=le.fit_transform(sleep_eff[f])
```

```python
sleep_eff.head()
```

```python
sleep_eff['Smoking status'].value_counts()
```

```python
X=sleep_eff.drop(columns='Smoking status',axis=1)
Y=sleep_eff['Smoking status']
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()

model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
pred=model.predict(X_test)
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(Y_test,pred)
```

```python
accuracy_score(Y_test,pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(X_train,Y_train)
pred=model.predict(X_test)
from sklearn.metrics import confusion_matrix,accuracy_score
print(confusion_matrix(Y_test,pred))
print(accuracy_score(Y_test,pred))
```

```python
input_data=(31,1,151,151,5,39,3,26,1,1	,0	,0	,0	,3)
input_data_as_numpy_array=np.asarray(input_data)
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)
prediction=model.predict(input_data_reshaped)
print(prediction)
if prediction[0]==0:
    print('the person have sleep efficiency')
else:
  print('the person have not sufficient sleep')
```

```python
from sklearn.svm import SVC
model=SVC(kernel='linear')
model.fit(X_train,Y_train)
pred=model.predict(X_test)
from sklearn.metrics import confusion_matrix,accuracy_score
print(confusion_matrix(Y_test,pred))
print(accuracy_score(Y_test,pred))
```
